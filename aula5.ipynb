````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

pd.set_option("display.precision", 2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
sns.set()
# Download dados_vendas_censo.csv
dados_vendas_censo = pd.read_csv("./dados_vendas_censo.csv")
dados_vendas_censo = dados_vendas_censo[
    [
    'Rua', 'Bairro', 'Cidade', 'Metragem', 'Quartos', 'Banheiros', 'Vagas', 
    'Valor_anuncio', 'Valor_mm', 'Valor_m2', 'latitude', 'longitude', 'Situacao_setor', 
    'V001', 'V002', 'V003', 'V004', 'V005' , 'V006', 'V007', 'V008', 'V009', 'V010', 'V011', 'V012'
    ]
].dropna()

dados_vendas_censo
````

````python
plt.figure(figsize=(10,8))
ax = sns.boxplot(data=dados_vendas_censo, y="Valor_anuncio")
plt.show()
````

````python
dados_vendas_censo.drop(dados_vendas_censo.query("Valor_anuncio > 30000000 | Valor_anuncio < 100000").index, inplace=True)
plt.figure(figsize=(10,8))
ax = sns.boxplot(data=dados_vendas_censo, y="Valor_anuncio")
plt.show()

````

````python
plt.figure(figsize=(10,8))
sns.scatterplot(data=dados_vendas_censo, x="Valor_anuncio", y="Valor_m2")
````

````python
dados_vendas_censo.corr()
````

````python
plt.figure(figsize=(20,8))
# read the article about correlations
mask = np.triu(np.ones_like(dados_vendas_censo.corr(), dtype=bool))
heatmap = sns.heatmap(dados_vendas_censo.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlação', fontdict={'fontsize':10}, pad=16)
````

````python
# Regressão linear (predizer valor continuo)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = dados_vendas_censo[["Metragem"]]
Y = dados_vendas_censo["Valor_anuncio"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=54)
lr = LinearRegression()
````

````python
X_train
````

````python
# treinar modelo
lr.fit(X_train, Y_train)
````

````python
# Predição
Y_predict_test = lr.predict(X_test)
````

````python
# resposta certa
Y_test
````

````python
from sklearn.metrics import mean_absolute_error, r2_score
mean_absolute_error(Y_test, Y_predict_test) # soma de todos os erros
````

````python
X = dados_vendas_censo[
    [
    'Metragem', 'Quartos', 'Banheiros', 'Vagas', 
    'latitude', 'longitude', 'Situacao_setor', 
    'V001', 'V002', 'V003', 'V004', 'V005' , 'V006', 'V007', 'V008', 'V009', 'V010', 'V011', 'V012'
    ]
]
Y = dados_vendas_censo["Valor_anuncio"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=54)
lr.fit(X_train, Y_train)
Y_predict_test = lr.predict(X_test)
mae_test = mean_absolute_error(Y_test, Y_predict_test)
r2 = r2_score(Y_test, Y_predict_test)
````

````python
mae_test # soma de todos os erros
````
